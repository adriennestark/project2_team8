# Tokenizer

1. Lowercase each word
2. Removing punctuation
3. Removing stopwords

```python
import re
from string import punctuation
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import WordNetLemmatizer, PorterStemmer
```

```python
# Instantiating the lemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
# Completing the tokenizer function
def tokenizer(text):
    """Tokenizes text."""
    # Expanding the default stopwords list if necessary
    sw = stopwords.words("english")
    sw.append("seen")
    sw.append("california")
    sw.append("file")
    sw.append("houston")
    sw.append("chars")
    sw.append("staff")
    sw.append("photo")
    sw.append("since")
    sw.append("taken")
    sw.append("illustration")
    sw.append("li")
    
    # cleaning text   
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub(' ', text)
    # Removing the punctuation
    words = word_tokenize(re_clean)
    # Converting the words to lowercase
    words = [word.lower() for word in words]
    # Removing the stop words
    words = [word for word in words if word not in sw]
    # Lemmatizing Words into root words
    tokens = [lemmatizer.lemmatize(word) for word in words]
    
    return tokens
```

```python
# Creating a new tokens column for lb
lb_df["tokens"] = lb_df["text"].apply(tokenizer)
lb_df.head(5)
```

```python
# Creating a new tokens column for ccl
ccl_df["tokens"] = ccl_df["text"].apply(tokenizer)
ccl_df.head(5)
```

# NGrams and Frequency Analysis

Look at the ngrams and word frequency for each ticker symbol

1. Set n-grams to N=2 using NLTK
2. List the top 20 words for each ticker

```python
from nltk import ngrams
from collections import Counter
```

```python
# Generating the lb N-grams where N=2!
lb_text = lb_df["text"].str.cat()
clean_lb_text =  tokenizer(lb_text)
bigram_counts = Counter(ngrams(clean_lb_text, n=2))
bigram_counts.most_common(20)
```

```python
# Generating the ccl N-grams where N=2!
ccl_text = ccl_df["text"].str.cat()
clean_ccl_text =  tokenizer(ccl_text)
bigram_counts = Counter(ngrams(clean_ccl_text, n=2))
bigram_counts.most_common(20)
```

```python
# Generating the dal N-grams where N=2!
dal_text = dal_df["text"].str.cat()
clean_dal_text =  tokenizer(dal_text)
bigram_counts = Counter(ngrams(clean_dal_text, n=2))
bigram_counts.most_common(20)
```

```python
# Generating the esty N-grams where N=2!
esty_text = esty_df["text"].str.cat()
clean_esty_text =  tokenizer(esty_text)
bigram_counts = Counter(ngrams(clean_esty_text, n=2))
bigram_counts.most_common(20)
```

```python
# Generating the sp500 N-grams where N=2!
sp500_text = sp500_df["text"].str.cat()
clean_sp500_text =  tokenizer(sp500_text)
bigram_counts = Counter(ngrams(clean_sp500_text, n=2))
bigram_counts.most_common(20)
```

```python
# Using the token_count function to generate the top 20 words
def token_count(tokens, N=20):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens).most_common(N)
```

```python
# Getting the top 20 words for lb
token_count(clean_lb_text, N=20)
```

```python
# Getting the top 20 words for ccl
token_count(clean_ccl_text, N=20)
```

```python
# Getting the top 20 words for dal
token_count(clean_dal_text, N=20)
```

```python
# Getting the top 20 words for esty
token_count(clean_esty_text, N=20)
```

```python
# Getting the top 20 words for sp500
token_count(clean_sp500_text, N=20)
```
